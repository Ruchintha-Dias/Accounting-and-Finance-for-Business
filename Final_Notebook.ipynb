## Predictions on Test dataset

```python
import pandas as pd
import joblib
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd

# Load the test dataset
test_data = pd.read_csv('/content/cleaned_test.csv')
print(test_data.head())
```

```python
# Load the saved encoder
ohe = joblib.load('outlet_city_encoder.joblib')

# Transform the categorical column using the loaded encoder
encode = ohe.transform(test_data[['outlet_city']])
# Create a dataframe
df_encoded = pd.DataFrame(encode, columns=ohe.get_feature_names_out(['outlet_city']))
test_data = test_data.join(df_encoded)

# Display first few rows
test_data.head()
```

```python
# Scaling

# Load the scaler
scaler = joblib.load("scaler.joblib")

numerical_columns = test_data[['luxury_sales', 'fresh_sales', 'dry_sales']]
X1 = scaler.fit_transform(numerical_columns)
```

```python
# Create a dataframe for Featuers (X) and Targets (y)
X = pd.concat([pd.DataFrame(X1, columns=['luxury_sales', 'fresh_sales', 'dry_sales']), df_encoded], axis=1)
```

```python
# Load the Best Model
best_lr = joblib.load("best_lr.joblib")
```

```python
predictions = best_lr.predict(X)  # Ensure test data has only feature columns
```

```python
Predicted_test_df = pd.DataFrame({ "Customer_ID": test_data["Customer_ID"],
                                   "Cluster_Category": predictions })  # predicted categories

# Save to CSV
Predicted_test_df.to_csv("Predicted_test.csv", index=False)
```
